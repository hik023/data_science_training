```python
import pandas as pd
import numpy as np
```

```python
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rc("font", size=18)
```

```python
df = pd.read_csv("1.4_houses.csv")
df.head(20)
```

```python
fig = plt.figure(figsize=(10, 10))

colors = {"basic": "grey", "medium": "green", "luxury": "gold"}
for d1, d2, l in zip(df.dim_1, df.dim_2, df.level):
    plt.scatter(d1, d2, color=colors[l], s=100)

plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
X = df[["dim_1", "dim_2"]]
```

```python
clust = KMeans(n_clusters=3).fit(X)
```

```python
[c1, c2, c3] = clust.cluster_centers_
```

```python
fig = plt.figure(figsize=(10, 10))

colors = {"basic": "grey", "medium": "green", "luxury": "gold"}
for d1, d2, l in zip(df.dim_1, df.dim_2, df.level):
    plt.scatter(d1, d2, color=colors[l], s=100)

plt.scatter(c1[0], c1[1], s=250, marker="x", c="black")
plt.scatter(c1[0], c1[1], s=250*5.3e2, alpha=0.1, c="black")
plt.scatter(c2[0], c2[1], s=250, marker="x", c="black")
plt.scatter(c2[0], c2[1], s=250*1e2, alpha=0.1, c="black")
plt.scatter(c3[0], c3[1], s=250, marker="x", c="black")
plt.scatter(c3[0], c3[1], s=250*2.5e2, alpha=0.1, c="black")

plt.show()
```
