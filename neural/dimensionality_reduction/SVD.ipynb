```python
import numpy as np
import matplotlib.pyplot as plt
import pickle

%matplotlib inline
```

```python
with open("data/eigen.pkl", "rb") as f:
    X = pickle.load(f)
```

```python
plt.plot(X[:, 0], X[:, 1], "x")
plt.axis("equal")
plt.show()
```

```python
from sklearn.decomposition import TruncatedSVD
```

```python
svd = TruncatedSVD(n_components=1).fit(X)

X_svd = svd.transform(X)
X_svd[:10]
```

```python
X_inv = svd.inverse_transform(X_svd)

plt.plot(X_inv[:, 0], X_inv[:, 1], "x")
plt.show()
```
