### 2.
Для каждой метрики найдите среднее, дисперсию, 50, 75 и 99 перцентили, моду. Какие выводы о наличии аномалий мы можем сделать?

```python
### YOUR CODE HERE ###
for col in df.columns:
    print(
        f"Column: {col}\n"
        f"Mean: {df[col].mean()}\n" 
        f"Dispersion: {df[col].std()}\n"
        f"50%: {df[col].quantile(0.5)}\n"
        f"75%: {df[col].quantile(0.75)}\n"
        f"99%: {df[col].quantile(0.99)}\n"
        f"Moda: {df[col].mode()}\n\n\n"

    )
```
