```python
%matplotlib inline
from sympy import *
from sympy.plotting import plot
import matplotlib.pyplot as plt
import numpy as np
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
x = Symbol("x")

def rmse(points, function):
    summ = 0
    for x_i, y_i in points:
        summ += (y_i - function.subs(x, x_i))**2
    mse = summ / 3
    return sqrt(mse)
```

```python
f = 1.1*x-110
f
```

```python
points = [(200,200), (450,300), (550,600)]
```

```python
rmse(points, f)
```

```python
plt.figure(3)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
x_arr = np.linspace(0, 600, num=1000)
plt.plot(x_arr, 1.1*x_arr-110)
x_s, y_s = [x for x, y in points], [y for x, y in points]
plt.scatter(x_s, y_s, c='r')
plt.show()
```
