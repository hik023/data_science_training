### f = x(y-5)<sup>2</sup>

---

```python
x, y = symbols("x, y")
```

```python
f1 = x * (y-5)**2
f1
```

```python
f1px = diff(f1, x)
f1px
```

```python
plot(f1px)
```

```python
solve(f1px)
```

```python
f1py = diff(f1, y)
f1py
```

```python
plot3d(diff(f1, y))
```

```python
solve(f1py)
```

MSE = sum((yi-f(x1i,x2i))<sup>2</sup>)/3

```python
a2, a1, a0 = symbols("a2, a1, a0")
```

```python
MSE = 1/3*((((1945*a1 + 560*a2 + a0)-4.3)**2) + \
           (((1495*a1 + 340*a2 + a0)-4.9)**2) + \
           (((1570*a1 + 343*a2 + a0)-5.2)**2))
MSE
```

```python
MSEpa2 = diff(MSE, a2)
MSEpa2
```

```python
MSEpa1 = diff(MSE, a1)
MSEpa1
```

```python
MSEpa0 = diff(MSE, a0)
MSEpa0
```

```python
nonlinsolve([MSEpa2, MSEpa1, MSEpa0], [a2,a1,a0])
```
