```python
a2, a1, a0 = symbols("a2,a1,a0")
```

```python
eq1_lp = a2 - a1 + a0 +1
eq2_lp = 4*a2 + 2*a1 + a0 - 8
eq3_lp = 4*a2 - 2*a1 + a0 + 8
```

```python
nonlinsolve([eq1_lp, eq2_lp, eq3_lp], [a2, a1, a0])
```
